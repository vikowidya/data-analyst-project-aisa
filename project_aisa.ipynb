```python
import pandas as pd

file_path = "/AISA_Financial_Analysis (1).xlsx"

# cek nama sheet
xls = pd.ExcelFile(file_path)
print(xls.sheet_names)
```

```python
df = pd.read_excel(file_path, sheet_name="Raw Data (Excel-Ready)")
df.head()
```

```python
df = pd.read_excel(file_path, sheet_name="Raw Data (Excel-Ready)", skiprows=1)
df.head()
```

```python
df.info()
```

```python
df['Tahun'] = pd.to_numeric(df['Tahun'], errors='coerce')
df = df.dropna(subset=['Tahun'])
df['Tahun'] = df['Tahun'].astype(int)
df.info()
display(df.head())
```

```python
df = df.copy()

df['Tahun'] = pd.to_numeric(df['Tahun'], errors='coerce')
df = df.dropna(subset=['Tahun'])

df['Tahun'] = df['Tahun'].astype(int)
```

```python
df['profit_margin'] = df['Net Profit'] / df['Revenue']
df['roe'] = df['Net Profit'] / df['Total Equity']
df['roa'] = df['Net Profit'] / df['Total Assets']
```

```python
plt.plot(df['Tahun'], df['Revenue'], marker='o', label='Revenue')
plt.plot(df['Tahun'], df['Net Profit'], marker='o', label='Net Profit')
plt.legend()
plt.title("Revenue vs Net Profit")
plt.show()
```

```python
plt.plot(df['Tahun'], df['profit_margin'], marker='o')
plt.title("Profit Margin Trend")
plt.show()
```
